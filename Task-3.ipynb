DATA VISUALIZATION

```python
# Overall Brief Histogram Plot 

data.hist(bins=20, figsize=(20,20), )
```

```python
# Pairplot over (Segment)

sns.pairplot(data, hue='Segment')
```

```python
# Heatmap (correlation)

sns.heatmap(correlation,annot=True)
```

```python
# Distributing Segment type over various inputs

data['Segment'].value_counts()
```

```python
# Plotting Pie chart for the above

plt.figure(figsize=(4,4),dpi=100)
plt.pie(data['Segment'].value_counts(), labels=data['Segment'].value_counts().index, autopct='%1.1f%%')
plt.title('SEGMENT')
plt.show()
```

```python
# Distributing Ship Mode type over various inputs

data['Ship Mode'].value_counts()
```

```python
# Plotting Countplot for the above

sns.countplot(data['Ship Mode'])
```

```python
# State wise Sales depiction

State_inspection=data.groupby('State')['Sales'].sum()
State_inspection
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=State_inspection.index,y=State_inspection.values, palette='rocket')
plt.ylabel("SALE")
plt.xticks(rotation=90)
plt.show()
```

```python
# State wise Profit depiction

State_overall=data.groupby('State')['Profit'].sum()
State_overall
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=State_overall.index,y=State_overall.values, palette='CMRmap')
plt.ylabel("PROFITS")
plt.xticks(rotation=90)
plt.show()
```

```python
# Comparing simultaneously (Sales and Profit)-->State

Analysis=data.groupby('State')['Sales','Profit'].sum()
Analysis.plot.bar(figsize=(15,10), ylabel='Sales')
```

```python
# Sub-Category Classification

objects=data.groupby('Sub-Category')['Profit'].sum()
plt.figure(figsize=(10,10))
sns.barplot(objects.index,objects.values)
plt.title('Object-Classification')
plt.xticks(rotation=90)
plt.show()
```

```python
# Discount Classfication

discount_area=data.groupby('State')['Discount'].sum()
plt.figure(figsize=(10,10))
sns.barplot(discount_area.index,discount_area.values)
plt.title('Discount-Classification')
plt.xticks(rotation=90)
plt.show()
```

```python
# Discount Dependency on Profit

sns.lineplot(x='Discount',y='Profit',data=data,color='blue')
plt.show()
```
